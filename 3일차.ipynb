```python
# 3번
s1 = '화면에 직접 출력'
s2 = 'ab\c'
s3 = 'does'
print(s1)
print(s2)
print(s3)
print(s1[0]) #[]를 사용하여 첫번째 단어 출력
print(s2[1]) #[]를 사용하여 두번째 단어 출력
print(s3[1:3]) #[]를 사용하여 두번째에서 3번째 단어 출력
print(s3[-1]) #뒤에서 첫번째 단어 출력
print(s3[-2]) #뒤에서 두번째 단어 출력
```

```python
# 4번
a = 2
b = 3.14
c = a+b
d = 1e2  #e2 : 10^2라는 뜻
e = 1e-2
print(a)
print(b)
print(c)
print(round(c,2)) #소수점 둘째자리까지 c를 출력
print("%.2f" % (c)) #소수점 둘째자리까지 c를 출력
print(a+b)
print(a,b,a+b,c)
print(d)
print(e)
```

```python
# 5번
item1 = '사과'
price1 = 1000
item2 = '바나나'
price2 = 500
print(item1, price1, sep=',',end='/') #sep로 두 값 사이에 ',' 출력, end로 개행문자를 /로 변경
print(item2, price2)
```

```python
# 6번
item1 = '사과'
price1 = 1000
item2 = '바나나'
price2 = 500
str1 = '{0}는 {1}원입니다.' #문자열에 {}를 사용하여 format을 통해 받을 값의 위치 지정
print(str1.format(item1, price1))
print(str1.format(item2, price2))
```

```python
# 8번
item1= '사과'
price1= 1000
item2= '바나나'
price2= 500
print(item1, price1, sep='*', end='/')
print(item2, price2)
```

```python
# 9번
item1= '사과'
price1= 1000
item2= '바나나'
price2= 500
str2 = '%s는 %d원입니다.'
print(str2% (item1, price1))
print(str2% (item2, price2))
```
